1. What is the difference between a function and a method in Python?
- Function: A function is defined using the def keyword and exists independently of any object. It can be called directly without being tied to a specific object.
- Functions are not bound to any object. They operate independently and do not require an instance of a class to be called.
- Functions are used for general-purpose tasks and can be written outside of any class.
- Called directly by its name (if in the same scope) or using the module name if imported

def greet(name):
    return f"Hello, {name}!"
print(greet("Ansh"))  
# Output: Hello, Ansh!

- Method: A method is essentially a function that is associated with an object (usually part of a class). It is called on an object and can access or modify the object's attributes.
- Methods are bound to an object or class. They typically take self (for instance methods) or cls (for class methods) as their first parameter to refer to the object or class they belong to
- Methods are used to define behaviors specific to a class or its instances.
- Called on an object or class using dot notation.

class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"
greeter = Greeter()
print(greeter.greet("Ansh"))  
# Output: Hello, Ansh!



2. Explain the concept of function arguments and parameters in Python.
- Parameters are the variables listed inside the parentheses in the function definition. They act as placeholders for the values (arguments) that will be passed to the function when it is called.

def greet(name):  # 'name' is a parameter
    print(f"Hello, {name}!")

- Arguments are the actual values or data you pass to a function when calling it. These values are assigned to the corresponding parameters.

greet("Alice")  # "Alice" is an argument



3. What are the different ways to define and call a function in Python?
- 3.1. Standard Function Definition and Call
Definition: Use the def keyword to define a function.
Call: Use the function name followed by parentheses

# Define
def greet(name):
    return f"Hello, {name}!"

# Call
print(greet("Ansh"))  # Output: Hello, Ansh!

- 3.2. Lambda (Anonymous) Functions
Definition: Use the lambda keyword for short, single-expression functions.
Call: Assign the lambda to a variable or use it directly.
# Define
square = lambda x: x ** 2

# Call
print(square(5))  # Output: 25

# Direct Call
print((lambda x: x + 10)(15))  # Output: 25

- 3.3 Functions with Default Arguments
Definition: Provide default values for parameters.
Call: Call with or without overriding the defaults.
# Define
def introduce(name, age=25):
    return f"My name is {name} and I am {age} years old."

# Call
print(introduce("Ansh"))  # Output: My name is Ansh and I am 25 years old.
print(introduce("Ashish", 40))  # Output: My name is Ashish and I am 40 years old.

- 3.4 Functions with Variable-Length Arguments
Definition: Use *args for positional arguments and **kwargs for keyword arguments.
Call: Pass multiple arguments dynamically.
# Define
def summarize(*args, **kwargs):
    return f"Args: {args}, Kwargs: {kwargs}"

# Call
print(summarize(1, 2, 3, name="Ansh", age=30))
# Output: Args: (1, 2, 3), Kwargs: {'name': 'Ansh', 'age': 30}

- 3.5 Nested Functions
Definition: Define a function inside another function.
Call: Call the inner function within the outer function or return it.
# Define
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

# Call
add_five = outer_function(5)
print(add_five(10))  
# Output: 15

- 3.6. Recursive Functions
Definition: A function that calls itself.
Call: Ensure a base case to avoid infinite recursion.
# Define
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

# Call
print(factorial(5))  
# Output: 120

- 3.7. Class Methods
Definition: Define functions inside a class using @staticmethod or @classmethod.
Call: Call via the class or an instance.
class Math:
    @staticmethod
    def add(a, b):
        return a + b

# Call
print(Math.add(3, 7))  
# Output: 10




4. What is the purpose of the `return` statement in a Python function?
- Send a Value Back to the Caller: It allows a function to send a result or value back to the part of the program that called it. For example:
def add(a, b):
    return a + b

result = add(3, 5)  # result will be 8
- End Function Execution: When a return statement is executed, the function terminates immediately, even if there are more lines of code after it.
For example:
def check_number(num):
    if num > 0:
        return "Positive"
    return "Non-positive"

print(check_number(5))  # Output: "Positive"

- Return Multiple Values:Python allows returning multiple values as a tuple, making it easy to return complex results. For example:
def get_coordinates():
    return 10, 20

x, y = get_coordinates()  # x = 10, y = 20

6. Explain the concept of generators in Python and how they are defined.
- Generators are a simple way to create iterators using functions and the yield keyword instead of returning values
- e.g. def countdown(n):
           while n >0:
               yield n
               n = n-1

       gen = countdown(5)  #Creating generator

       for num in gen:         #using the generator
           print(num)



7. What are the advantages of using generators over regular functions?

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            return False
    return True

for num in range(1,200):
  if is_prime(num):
    print(f"{num} is a PRIME number")
  else:
    print(f"{num} is not a prime number")
```

```python
# 11.Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity.
# The product should be increased by 10,- € if the value of the order is smaller than 100,00 €. Write a Python program using lambda and map


orders = [
    (34587, 40.95, 4),  # Example order 1
    (98762, 56.80, 5),  # Example order 2
    (77226, 32.95, 3),  # Example order 3
    (88112, 24.99, 3)  # Example order 4
]

process_order = lambda order: (
    order[0],
    order[1] * order[2] + (10 if order[1] * order[2] < 100 else 0)
)


result = list(map(process_order, orders))

print(result)
```
